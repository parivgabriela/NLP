# Analogías

En este notebook vamos a desarollar un ejemplo práctico del uso de similitud de vectores. Se trabajará  con un dataset de vectores de palabras preentrenado. Con la implementación de la librería gensim vamos a utilizar los embeddings para luego calcular las similitudes, como la del coseno, esto implica que las palabras doctor y enfermera estarán mucho más cercanas que doctor y nube.

## Descargar e importar dataset de Kaggle
link al dataset https://www.kaggle.com/api/v1/datasets/download/rtatman/pretrained-word-vectors-for-spanish?dataset_version_number=1

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("rtatman/pretrained-word-vectors-for-spanish")

print("Path to dataset files:", path)
```

## Cargar el modelo word2vec

```python
import gensim

vectores = gensim.models.KeyedVectors.load_word2vec_format(path+'/SBW-vectors-300-min5.txt')
```

```python
def analogias(palabra1, palabra2, palabra3):
  similitud = vectores.most_similar(positive=[palabra1, palabra3], negative=[palabra2] )
  print(f"{palabra1} es a {palabra2} como {similitud[0][0]} es a {palabra3}")
```

## Demo de analogías

Vamos a probar como se comporta el modelo cuando le pasamos tres parámetros y que nos responda en base a las palabras ingresadas, vamos a ver que algunas palabras las entiende bastante bien y responde acorde, pero en el ejemplo de 'carnívoro', 'carne', 'herbívoro' responde de forma incorrecta. Esto puede deberse a como esta entrenado el modelo.

```python
analogias('rey', 'hombre', 'mujer')
analogias('Alemania', 'alemán', 'español')
analogias('pan', 'trigo', 'leche')
```

```python
analogias('carnívoro', 'carne', 'herbívoro')
analogias('Madrid', 'España', 'Italia')
```

## Demo Cercanos

Vamos definir la función de cercanos para que dad una palabra me devuelva los similares, quellas palabras que esten relacionadas. En el caso de Argentina vs argentina, entendio el contexto y pudo definir que hace referncia a la nacionalidad en lugar del país. A tomar en cuenta para casos es los que el texto debe identificar las palabras de un texto ingresado por un usuario, ya que al escribir se suelen cometer errores ortográficos.

```python
def cercanos(palabra):
    vecinos = vectores.most_similar(positive=[palabra])
    print(f'vecinos de: {palabra}')
    for word, score in vecinos:
        print("\t%s"%word)
```

```python
cercanos("rey")
```

```python
cercanos("Argentina")
```

```python
cercanos("argentina")
```
